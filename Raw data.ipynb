```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
#Importing data 
data = pd.read_csv('Movie data.csv')
```

```python
#let's look at some of the rows
data.sample(n=10)
```

```python
data.columns
```

```python
# check if there are missing values
data.isna().sum()
data = data.dropna() #For now, I am dropping all of the missing values
data.isna().sum()
```

```python
data['revenue'].shape
```

```python
data['runtime'].shape
data.sample(n=10)
```

# Check the distribution of the main variables



```python
plt.figure(figsize=(20,10))
plt.scatter(data['runtime'],data['vote_average'])
plt.xlabel('Run Time',fontsize=20)
plt.ylabel('Movie Rating',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data['revenue'],data['vote_average'])
plt.xlabel('Revenue',fontsize=20)
plt.ylabel('Movie Rating',fontsize=20)
plt.show()
```

```python
# repeat the scatter plot with log of revenue
data['log_revenue'] = np.log10(data['revenue'])
plt.figure(figsize=(20,10))
plt.scatter(data['log_revenue'],data['vote_average'])
plt.xlabel('Log of Revenue',fontsize=20)
plt.ylabel('Movie Rating',fontsize=20)
plt.show()
```

```python
data['runtime'].unique()
```

# Check the distribution of the main variable of interest

```python
plt.figure(figsize=(10,7))
data['vote_average'].hist(bins=10, alpha=0.5,grid=False)
plt.title('Histogram of Values',fontsize=20)
plt.xlabel('Values',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,7))
data['revenue'].hist(alpha=0.5,grid=False)
plt.title('Histogram of Values',fontsize=20)
plt.xlabel('Values',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.show() 
```

 The above graph shows that the revenue is highly skewed.

```python
data['original_language'].unique()
```

Let's compare the english language movies with non-english movies.

```python
data['is_english'] = data['original_language'].apply(lambda x: 1 if x == 'en' else 0)
```

```python
data[['original_language','is_english']]
```

```python
plt.figure(figsize=(200,10))

sns.pairplot(data,x_vars=['log_revenue'],
            y_vars=['vote_average'],hue='is_english',height=7)
plt.xlabel('Log of Revenue')
plt.ylabel('Movie Ratings')
plt.show()
```
